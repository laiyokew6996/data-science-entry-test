```python
def swap(x, y):
    """
    Function: swap
    ----------------
    This function swaps the values of x and y.

    Requirements:
    1. x and y must be numeric (int or float).
    2. If either value is not numeric, return -1.
    3. If both are numeric, print the swapped values.
    """

    # Step 1: Check whether both x and y are numeric
    # isinstance() is used to verify the data type
    if not (isinstance(x, (int, float)) and isinstance(y, (int, float))):
        return -1  # Return -1 if any value is not numeric

    # Step 2: Swap the values using only x and y
    # Python allows swapping without a temporary variable
    x, y = y, x

    # Step 3: Print the swapped values
    print("Swapped values:", x, y)


# =========================
# Task 2: Function Invocation
# =========================

# Scenario 1: One value is not numeric

result1 = swap("Apple", 10)
print(result1)

# Scenario 2: Both values are numeric
# Expected: Function should print the swapped values
swap(9, 17)
```
