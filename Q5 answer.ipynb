```python
def check_divisibility(num, divisor):
    """
    Check whether a number is divisible by another number.

    Parameters:
    num (int or float): The number to be checked.
    divisor (int or float): The number used to divide.

    Returns:
    bool: 
        True  -> if num is divisible by divisor
        False -> if not divisible or invalid input
    """

    # Step 1: Validate input types
    # Ensure both inputs are numeric (integer or float)
    if not isinstance(num, (int, float)) or not isinstance(divisor, (int, float)):
        print("Invalid input: Both values must be numeric.")
        return False

    # Step 2: Prevent division by zero
    # Division by zero is not allowed in mathematics or Python
    if divisor == 0:
        print("Invalid input: Divisor cannot be zero.")
        return False

    # Step 3: Check divisibility
    # The modulus (%) operator returns the remainder after division
    remainder = num % divisor

    # If the remainder is zero, the number is divisible
    if remainder == 0:
        return True
    else:
        return False


# -----------------------------
# Task 2: Test the function
# -----------------------------

# Scenario 1
result1 = check_divisibility(10, 2)
print("Is 10 divisible by 2?", result1)

# Scenario 2
result2 = check_divisibility(7, 3)
print("Is 7 divisible by 3?", result2)
```
