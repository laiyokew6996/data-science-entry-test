```python
def find_first_negative(lst):
    """
    This function finds the first negative number in a list.

    Parameters:
    lst (list): A list of numbers.

    Returns:
    int or str:
        - The first negative number if found.
        - "No negatives" if there is no negative number in the list.
    """

    # Step 1: Start from the first index of the list
    index = 0  

    # Step 2: Use a while loop to check each element
    # The loop continues as long as the index is within the list length
    while index < len(lst):

        # Step 3: Check if the current element is negative
        if lst[index] < 0:
            return lst[index]  # Return the first negative number found

        # Step 4: Move to the next element
        index += 1

    # Step 5: If no negative number is found after checking all elements
    return "No negatives"
```

```python
# Scenario 1: List contains negative numbers
list1 = [3, 5, -1, 7, -2, 8]
result1 = find_first_negative(list1)
print("First negative in list1:", result1)

# Scenario 2: List has no negative numbers
list2 = [2, 10, 7, 0]
result2 = find_first_negative(list2)
print("First negative in list2:", result2)
```
