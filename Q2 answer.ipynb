```python
def find_and_replace(lst, find_val, replace_val):
    """
    Function: find_and_replace
    ----------------------------------------
    This function searches for all occurrences of a value (find_val)
    in a given list (lst) and replaces them with another value (replace_val).

    Parameters:
    lst (list): The list in which the search and replacement will occur.
    find_val : The value to search for in the list.
    replace_val : The value to replace the found elements with.

    Returns:
    list: The modified list after replacement.
    """

    # Step 1: Validate input
    # Ensure the input is a list to prevent runtime errors.
    if not isinstance(lst, list):
        return "Error: Input must be a list."

    # Step 2: Loop through the list using index
    # This allows us to modify elements directly.
    for index in range(len(lst)):

        # Step 3: Check if the current element matches the value to find
        if lst[index] == find_val:

            # Step 4: Replace the matched value
            lst[index] = replace_val

    # Step 5: Return the updated list
    return lst


# ----------------------------
# Task 2: Function Invocation
# ----------------------------

# Scenario 1: Replace all 2s with 5
result1 = find_and_replace([1, 2, 3, 4, 2, 2], 2, 5)
print("Scenario 1 result:", result1)

# Scenario 2: Replace "apple" with "orange"
result2 = find_and_replace(["apple", "banana", "apple"], "apple", "orange")
print("Scenario 2 result:", result2)
```
