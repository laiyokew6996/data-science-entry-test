```python
def string_reverse(s):
    """
    Function: string_reverse
    
    Purpose:
    Reverse a given string.

    Parameters:
    s (str) : The input value that must be a string.

    Returns:
    str : The reversed string.
    If the input is not a string, a message is returned.
    """

    # Step 1: Validate that the input is a string
    if not isinstance(s, str):
        return "Error: Input must be a string."

    # Step 2: Initialize an empty string to store the reversed result
    reversed_string = ""

    # Step 3: Loop through the string from the last character to the first
    # range starts from the last index and moves backward
    for i in range(len(s) - 1, -1, -1):
        reversed_string += s[i]

    # Step 4: Return the final reversed string
    return reversed_string


# Task 2: Invoke the function with the given scenarios

# Scenario 1
result1 = string_reverse("Hello World")
print(result1)

# Scenario 2
result2 = string_reverse("Python")
print(result2)
```
