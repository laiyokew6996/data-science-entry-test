```python
# ---------------------------------------------------------
# Car Class Definition
# ---------------------------------------------------------

class Car:
    """
    The Car class represents a car object.

    Each car has three main attributes:
    - make  : the manufacturer of the car
    - model : the model name
    - year  : the production year

    The class also includes a method to display
    the car details in a structured format.
    """

    def __init__(self, make, model, year):
        """
        Constructor method used to initialise a Car object.

        Parameters:
        make (str)  : brand of the car (e.g., Toyota)
        model (str) : model of the car (e.g., Corolla)
        year (int)  : year the car was produced
        """

        # Store the input values as object attributes
        self.make = make
        self.model = model
        self.year = year

    def describe_car(self):
        """
        This method prints the car information.

        It combines the year, make, and model into one
        readable sentence using an f-string.
        """

        # Display car details in the required format
        print(f"{self.year} {self.make} {self.model}")


# ---------------------------------------------------------
# Task 2: Create an instance and test the class
# ---------------------------------------------------------

# Create a Car object with the given details
car1 = Car("Toyota", "Corolla", 2020)

# Call the method to display the car information
car1.describe_car()
```
