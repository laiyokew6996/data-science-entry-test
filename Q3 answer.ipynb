```python
def update_dictionary(dct, key, value):
    """
    Function: update_dictionary

    Purpose:
    Update a dictionary by adding or modifying a key-value pair.

    Parameters:
    dct (dict) : The dictionary to be updated.
    key        : The key to be added or updated.
    value      : The value to assign to the key.

    Behaviour:
    - If the key already exists, the function prints the original value.
    - The function then updates the dictionary with the new value.

    Returns:
    dict : The updated dictionary.
    """

    # Step 1: Validate that the input 'dct' is a dictionary
    if not isinstance(dct, dict):
        return "Error: First argument must be a dictionary."

    # Step 2: Check whether the key already exists in the dictionary
    if key in dct:
        # Display the original value before updating
        print("Original value:", dct[key])

    # Step 3: Add or update the key-value pair in the dictionary
    dct[key] = value

    # Step 4: Return the updated dictionary
    return dct


# Task 2: Invoke the function with the given scenarios

# Scenario 1: Add a new key-value pair
result1 = update_dictionary({}, "name", "Alice")
print(result1)

# Scenario 2: Update an existing key-value pair
result2 = update_dictionary({"age": 25}, "age", 26)
print(result2)
```
